```python
import cv2
from engineering_notation import EngNumber  as eng # only from pip
cv2.namedWindow('dvs',cv2.WINDOW_NORMAL)
histrange = [(0, v) for v in (IMSIZE, IMSIZE)]  # allocate DVS frame histogram to desired output size
i=0
vflow_ppus=5e-3 # estimate vertical flow
EVENT_COUNT_CLIP_VALUE=4
while True:
    if i>len(data)-EVENT_COUNT_PER_FRAME:
        i=0
    elif i<0:
        i=len(data)-EVENT_COUNT_PER_FRAME
    tts=ts[i:i+EVENT_COUNT_PER_FRAME]
    xxs=xs[i:i+EVENT_COUNT_PER_FRAME]
    yys=ys[i:i+EVENT_COUNT_PER_FRAME]
    dt=tts-tts[0]
    yys=yys-vflow_ppus*dt
    frame, _, _ = np.histogram2d(yys, xxs, bins=(IMSIZE, IMSIZE), range=histrange)
    fmax_count=np.max(frame)
    frame[frame > EVENT_COUNT_CLIP_VALUE]=EVENT_COUNT_CLIP_VALUE
    frame=frame/EVENT_COUNT_CLIP_VALUE # max pixel will have value 255
    cv2.imshow('dvs',frame)
    k=cv2.waitKey(0)
    if k==ord('x'):
        cv2.destroyAllWindows()
        break
    elif k==ord('f'):
        i+=100*EVENT_COUNT_PER_FRAME
    elif k==ord('b'):
        i-=EVENT_COUNT_PER_FRAME
    elif k==ord('i'):
        vflow_ppus*=1.2
        print(f'\rvflow_ppus={eng(vflow_ppus)}',end='')
    elif k==ord('d'):
        vflow_ppus/=1.2
        print(f'\rvflow_ppus={eng(vflow_ppus)}',end='')
    elif k==ord('c'):
        vflow_ppus=-vflow_ppus
        print(f'\rvflow_ppus={eng(vflow_ppus)}',end='')
    elif k==ord(' '):
        i+=EVENT_COUNT_PER_FRAME




```
